**Question 1: How much more playing time did the player with the highest playing time have in Real Madrid than in Paris Saint Germain?**

*Table 1: Real Madrid*

Web-Scraping/Requesting the URL HTML data from the given URL for Real Madrid from FBREF.com.

```python
url_1 = 'https://fbref.com/en/squads/53a2f082/Real-Madrid-Stats'
response = requests.get(url_1)
status = response.status_code
if status == 200:
    page = response.text
    soup = bs(page)
else:
    print(f"Oops! Received status code {status}")
```

Parsing the data using BeautifulSoup.

```python
print(soup.prettify())
type(soup)
```

Finding the table the I want to use to answer the question and putting it into a pandas data frame, not including the first header (there is a two tiered header and I removed the first tier).

```python
table_1 = soup.find(class_="stats_table sortable min_width")
table_1_df = pd.read_html(str(table_1), skiprows=[0], header=0)[0]
table_1_df
```

Cleaning the data by dropping all the columns that I do not need to answer the question, only including the 'Player' and 'Min' columns.

```python
table_1_df2 = table_1_df.drop(axis=1, columns=['Nation', 'Pos', 'Age', 'MP', 'Starts', '90s', 'Gls', 'Ast', 'Ast.1', 'G+A.1', 'G-PK.1', 'G+A-PK', 'xG.1', 'xAG.1', 'xG+xAG', 'npxG.1', 'npxG+xAG.1', 'Matches', 'G+A', 'G-PK', 'PK', 'PKatt', 'CrdY', 'CrdR', 'xG', 'npxG', 'xAG', 'npxG+xAG', 'PrgC', 'PrgP', 'PrgR', 'Gls.1'])
table_1_df2
```

Removing the row that has the value 'Opponent Total' in the Player column since I do not want to include it in the answer.

```python
value_to_remove = 'Opponent Total' 
table_1_df3 = table_1_df2[table_1_df2['Player'] != value_to_remove]
table_1_df3
```

Removing the row that has the value 'Squad Total' in the Player column since I do not want to include it in the answer.

```python
value_to_remove_2 = 'Squad Total'
table_1_df4 = table_1_df3[table_1_df3['Player'] != value_to_remove_2]
table_1_df4
```

Finding and displaying the row in the cleaned data frame that has the highest value in the 'Min' column to get the player in Real Madrid that had the highest playing time in minutes.

```python
highest_playing_time = table_1_df4['Min'].max()

player_with_highest_time = table_1_df4[table_1_df4['Min'] == highest_playing_time]
player_with_highest_time
```

**Question 5: How many players of each nationality are in the 2023-2024 Real Madrid and Paris Saint Germain teams?**

*Table 1: Real Madrid*

Grouping the first table from url_1 by 'Nation' and aggregating the 'Player' column to find the count of players from each nationality. The reset_index() function makes output display as a table.

```python
table_7_df = table_1_df.groupby('Nation')['Player'].count()
table_7_df2 = table_7_df.reset_index()
table_7_df2
```

**Answer: There is 1 player from Argentina, 1 player from Austria, 4 players from Brazil, 2 players from Germany, 1 player from England, 18 players from Spain, 4 players from France, 1 player from Croatia, 1 player from Morocco, 1 player from Turkey, 1 player from Ukraine, and 2 players from Uruguay in Real Madrid in the 2023-2024 Season.**

Making a barplot using seaborn of the number of players of each nationality in Real Madrid in the 2023-2024 season. The x-labels were renamed and they were rotated to make them more easily readible.

```python
x_labels_1 = ['Argentina', 'Austria', 'Brazil', 'Germany', 'England', 'Spain', 'France', 'Croatia', 'Morocco', 'Turkey', 'Ukraine', 'Uruguay']

sns.barplot(x='Nation', y='Player', data=table_7_df2)
plt.xticks(ticks=range(len(x_labels_1)), labels=x_labels_1, rotation=45) 

plt.xlabel('Nationalities')
plt.ylabel('Number of Players')
plt.title('Number of Players of Each Nationality in Real Madrid in 2023-2024 Season')
plt.show()
```
